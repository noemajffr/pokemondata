```python
#Importation des librairies
import pandas as pd
```

```python
#Lecture du fichier
df = pd.read_csv("Data.csv", index_col="Pokemon Id")
pd.set_option("max_columns", None)
```

On commence par créer les données nécéssaires pour programmer le mini-jeu.

```python
#Sélection des formes originales
df = df[df["Alternate Form Name"].isna()]
```

```python
#Data frame composé des starters uniquement
df_starters = df.iloc[[0,3,6,
                       151,154,157,
                       251,254,257,
                       386,389,392,
                       494,497,500,
                       649,652,655,
                       721,724,727,
                       809,812,815]]
```

```python
def signe_utilisateur(jour_de_naissance,mois_de_naissance) :
    """Trouve le signe astrologique de l'utilisateur"""
    
    #Définition d'un dictionnaire
    astrologie = {
    "Bélier" : (21,3,20,4),
    "Taureau" : (21,4,21,5),
    "Gémeaux" : (22,5,21,6),
    "Cancer" : (22,6,22,7),
    "Lion" : (23,7,22,8),
    "Vierge" : (23,8,22,9),
    "Balance" : (23,9,22,10),
    "Scorpion" : (23,10,22,11),
    "Sagittaire" : (23,11,21,12),
    "Capricorne" : (22,12,20,1),
    "Verseau" : (21,1,19,2),
    "Poissons" : (20,2,20,3)
    }
    
    #Définition d'une boucle conditionnelle
    for signe, date in astrologie.items() :
        if ((jour_de_naissance>=date[0]) and (mois_de_naissance==date[1])) or ((jour_de_naissance<=date[2]) and (mois_de_naissance==date[3])) :
            return(signe)
```

```python
def type_utilisateur(signe_utilisateur) :
    """Convertit le signe astrologique de l'utilisateur en type élémentaire"""
    
    #Définition d'un dictionnaire
    astrologie_type = {
    "Grass" : ("Gémeaux","Vierge","Balance","Capricorne"),
    "Water" : ("Cancer","Scorpion","Verseau","Poissons"),
    "Fire" : ("Bélier","Taureau","Lion","Sagittaire")
    }
    
    #Définition d'une boucle conditionnelle
    for types, signes in astrologie_type.items() :
        for signe in signes :
            if (signe_utilisateur==signe) :
                return(types)
```

```python
def region_utilisateur(region_de_naissance) :
    """Convertit la région de l'utilisateur en région Pokemon"""
    
    #Définition d'un dictionnaire
    regions = {
        "Kanto" : ("Ile-de-France","Bretagne"),
        "Johto" : "Normandie",
        "Hoenn" : ("Occitanie","Pays de la Loire"),
        "Sinnoh" : "Nouvelle-Aquitaine",
        "Unova" : ("Provence-Alpes-Côte d’Azur","Centre-Val de Loire"),
        "Kalos" : ("Hauts-de-France","Grand Est"),
        "Alola" : "Corse",
        "Galar" : ("Auvergne-Rhône-Alpes","Bourgogne-Franche-Comté")
    }
    
    #Définition d'une boucle conditionnelle
    for regions_pokemon, regions_france in regions.items() :
        if (len(regions_france)<=2) :
            for region in regions_france :
                if (region_de_naissance==region) :
                    return(regions_pokemon)
        else :
            if (region_de_naissance==regions_france) :
                    return(regions_pokemon)
```

```python
def pokemon_utilisateur() :
    """Convertit les critères de l'utilisateur en un starter Pokemon"""
    
    #Entrée des données utilisateur
    jour_de_naissance = int(input("Jour de naissance : "))
    mois_de_naissance = int(input("Mois de naissance : "))

    #Appel de deux fonctions
    signe = signe_utilisateur(jour_de_naissance,mois_de_naissance)
    print(f"Ton signe : {signe}")
    types = type_utilisateur(signe)
    print(f"\nTon type : {types}")
    
    #Tri du dataframe selon les données utilisateur
    df_select = df_starters[df_starters["Primary Type"]==types]
    regions = sorted(["Ile-de-France","Bretagne","Normandie","Occitanie",
               "Pays de la Loire","Nouvelle-Aquitaine","Provence-Alpes-Côte d’Azur","Centre-Val de Loire",
               "Hauts-de-France","Grand Est","Corse","Auvergne-Rhône-Alpes","Bourgogne-Franche-Comté"])
    print("\nRégion de naissance :")
    region_de_naissance = None
    while region_de_naissance not in regions :
        region_de_naissance = input(f"\n{regions}\nVotre réponse : ")
    region = region_utilisateur(region_de_naissance)
    print(f"\nTa région : {region}")
    df_select = df_select[df_select["Region of Origin"]==region]
    
    #Affichage du Pokemon
    nom_du_pokemon = df_select["Pokemon Name"]
    print(f"\nTon Pokemon : {nom_du_pokemon}")
```

### Vous pouvez ainsi découvrir quel est votre starter en faisant le test ci-dessous : 

```python
#Appel d'une fonction
pokemon_utilisateur()
```
